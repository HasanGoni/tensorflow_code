```python
import tensorflow as tf
```

```python
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
Path.ls = lambda x: list(x.iterdir())
```

```python
path = Path(r'/tmp/fcnn/dataset1/annotations_prepped_test')
```

```python
annotation = tf.image.decode_jpeg(tf.io.read_file(path.ls()[0].as_posix()))
```

```python
plt.imshow(annotation)
```

```python
ano_resize = tf.image.resize(annotation, (224, 224,))
```

```python
plt.imshow(ano_resize)
```

```python
int_ano = tf.cast(ano_resize, dtype=tf.int32)
```

```python
plt.imshow(int_ano)
```

```python
annotation = tf.reshape(int_ano, (224, 224, 1,))
```

```python
plt.imshow(annotation)
```

```python
plt.imshow(tf.equal(annotation[:,:,0], tf.constant(1)))
```

```python
plt.imshow(tf.equal(annotation[:,:,0], tf.constant(0)))
```

```python
plt.imshow(tf.equal(annotation[:,:,0], tf.constant(2)))
```

```python
plt.imshow(tf.equal(annotation[:,:,0], tf.constant(3)))
```

```python
plt.imshow(tf.equal(annotation[:,:,0], tf.constant(4)))
```

```python
plt.imshow(tf.equal(annotation[:,:,0], tf.constant(5)))
```
